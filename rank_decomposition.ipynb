```python
import torch
import numpy as np
_ = torch.manual_seed(0)
```

Gererate a rank deficient matrix W

```python
d, k = 10, 10

W_rank = 2
W = torch.randn(d,W_rank) @ torch.randn(W_rank,k)
W
```

Rank Evaluation

```python
W_rank = np.linalg.matrix_rank(W)
print(f"Rank of W: {W_rank}")
```

Calculating Singular Value Decomposition

```python
#perform SVD on W (W=UXSXV^T)
U, S, Vt = torch.linalg.svd(W)

#For rank-r factorization, take first r columns of U and Vt, and first r singular values from S

U_r = U[:, :W_rank]
S_r = torch.diag(S[:W_rank])
V_r = Vt[:,  :W_rank].t() #transpose Vt to get V


#Compute B = U_r @ S_r and A = V_r 
B = U_r @ S_r
A = V_r
print(f'Shape of B: {B.shape}')
print(f'Shape of A: {A.shape}')
```

```python
# Generate random bias and input
bias = torch.randn(d)
x = torch.randn(d)

# Compute y = Wx + bias
y = W @ x + bias
# Compute y' = (B*A)x + bias
y_prime = (B @ A) @ x + bias

print("Original y using W:\n", y)
print("")
print("y' computed using BA:\n", y_prime)
```

```python
print("Total parameters in W:", W.numel())
print("Total parameters in B and A:", B.numel() + A.numel())
```
